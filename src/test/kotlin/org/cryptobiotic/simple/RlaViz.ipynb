```kotlin
import java.security.SecureRandom

/* data class CVR(val id: Int, val vote: Boolean)

val random = SecureRandom.getInstanceStrong()

fun generateCVRs(n: Int): List<CVR> {
    val result = mutableListOf<CVR>()
    repeat(n) { result.add(CVR(n, random.nextBoolean()))}
    return result
}

 */

fun runTest(ntrials: Int, nballots: Int, show: Boolean = false, plot: Boolean = false): List<Pair<Double, Double>> {
    val hist = mutableListOf<Pair<Double, Double>>()
    repeat(ntrials) {
        val cvrs = generateCVRs(nballots)
        val votes = cvrs.map { if (it.vote) 1.0 else 0.0 }.sum()
        val pct = votes/nballots
        hist.add(Pair(votes, pct))
        if (show) println("trial $it count ${votes/nballots}")
    }
    return hist
}

val ntrials = 200
val nballots = 1000
val hist = runTest(ntrials, nballots)
```

```kotlin
val pct = hist.map { it.second }
val dict = mapOf("pct" to pct )
val histPlot = plot(dict) {
    histogram(pct)

    layout.title = "pct voted for"
}
histPlot
```

```kotlin
val histd = DoubleArray(pct.size) { pct[it] }
val stats = org.jetbrains.kotlinx.statistics.math3.stat.descriptive.DescriptiveStatistics(histd)
print("ntrial $ntrials nballots $nballots:")
print("  nvotes = ${stats.mean} ")
print("  mean = ${stats.mean} ")
print("  variance = ${stats.variance} ")
print("  standardDeviation = ${stats.standardDeviation} ")
print("  skewness = ${stats.skewness} ")
println()
```

```kotlin
val votes = hist.map { it.first }
val dictv = mapOf("votes" to votes )
val histPlotV = plot(dictv) {
    histogram(votes)
    layout.title = "votes per 1000"
}
histPlotV
```
